# Introduction

## What is a Histogram?

(A bar plot shows how many in each category)


```python
## NOTE: No need to edit
(
    df_mpg
    >> gr.ggplot(gr.aes("class"))
    + gr.geom_bar()
)
```

(Continuous variable)

```python
## NOTE: No need to edit
(
    df_mpg
    >> gr.ggplot(gr.aes("displ"))
    + gr.geom_bar()
)
```

(Create groups of values)

```python
## NOTE: No need to edit
(
    df_mpg
    >> gr.tf_mutate(displ_group=DF.displ // 1)
    >> gr.tf_count(DF.displ_group)
)
```

(Visualize the count within each group.)

```python
## NOTE: No need to edit
(
    df_mpg
    >> gr.tf_mutate(displ_group=DF.displ // 1)
    >> gr.tf_count(DF.displ_group)
    >> gr.ggplot(gr.aes("displ_group", "n"))
    + gr.geom_col()
)
```

(The geometry `gr.geom_histogram()` does this automatically.)

```python
(
    df_mpg
    >> gr.ggplot(gr.aes("displ"))
    + gr.geom_histogram(bins=20)
)
```
